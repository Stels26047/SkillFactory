```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
path_driver = 'C:/Users/sidor/Downloads/chromedriver.exe'
def run(path_driver):
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_experimental_option("prefs", {"profile.default_content_setting_values.cookies": 2})
    chrome_options.add_argument("--incognito")
    driver = webdriver.Chrome(executable_path=path_driver, chrome_options=chrome_options)
    driver.get("https://tproger.ru/quiz/real-programmer/")
    button = driver.find_element(By.XPATH,'//*[@id="start_quiz"]')
    button.click()
    
    index_XPATH = 255
    index = 1
    stop = 1
    while stop != 13:
        
        question = driver.find_element(By.XPATH, '//*[@id="quiz_item_{}"]/div[1]/p'.format(index_XPATH)).text
        answers = driver.find_elements(By.CLASS_NAME, "answer")
        correct_answer = driver.find_element(By.XPATH,'//*[@id="quiz_item_{}"]/div[2]/div[1]'.format(index_XPATH))
        print('Вопрос',index,question)
        for ans in answers: 
            if ans.text:
                correct_answer.click()
                print()
                print(ans.text)
        button = driver.find_element(By.XPATH,'//*[@id="quiz_item_{}"]/button'.format(index_XPATH))
        button.click()
        index_XPATH +=1
        index +=1
        stop += 1
        print()
        print()
    driver.close()
    return driver
```

```python
run(path_driver)
```
